### **Irish Dataset Plotting**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_names'] = iris_df['target'].apply(lambda x: iris.target_names[x])
```

```python
iris_df.head()
```

## **Bar Plotting using Mean**

```python
# group the data by Mean of Species
grouped_iris_df_mean = iris_df.groupby('target_names').mean()
```

```python
# transposing rows and column for easy plotting
# features = x-axis, mean_value = y-axis
grouped_iris_df_mean_T = grouped_iris_df_mean.transpose()
```

```python
grouped_iris_df_mean_T.plot(kind='bar', figsize=(10, 6))
plt.ylabel('Mean Value')
plt.xlabel('Features')
plt.legend(title='Species')
```

## **Scatter plotting using Median**

```python
# Grouped the Data by Median of species
grouped_iris_df_median = iris_df.groupby('target_names').median()
# grouped_iris_df_median.head()
```

```python
# Create the Scatter Plot
# Extracting median values for scatter plot
species = grouped_iris_df_median.index
sepal_length_median = grouped_iris_df_median['sepal length (cm)']
sepal_width_median = grouped_iris_df_median['sepal width (cm)']
petal_length_median = grouped_iris_df_median['petal length (cm)']
petal_width_median = grouped_iris_df_median['petal width (cm)']
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(sepal_length_median, sepal_width_median, label='Sepal Length vs Sepal Width', color='r')
plt.scatter(petal_length_median, petal_width_median, label='Petal Length vs Petal Width', color='b')

plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.legend()
```

## **Box Plotting**

```python
# Create the Box Plots using Seaborn
plt.figure(figsize=(12, 10))
```

```python
# Box plot for Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(x='target_names', y='sepal length (cm)', data=iris_df)
plt.title('Sepal Length (cm)')

# Box plot for Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(x='target_names', y='sepal width (cm)', data=iris_df)
plt.title('Sepal Width (cm)')

# Box plot for Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(x='target_names', y='petal length (cm)', data=iris_df)
plt.title('Petal Length (cm)')

# Box plot for Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(x='target_names', y='petal width (cm)', data=iris_df)
plt.title('Petal Width (cm)')

# Adjust layout
plt.tight_layout()
```
